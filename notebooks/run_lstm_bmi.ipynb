# Basic Model Interface (BMI) for streamflow prediction using Long Short-Term Memory (LSTM) networks
This Long Short-Term Memory (LSTM) network was developed for use in the [Next Generation National Water Model (Nextgen)](https://github.com/NOAA-OWP/ngen). Nextgen runs models with [Basic Model Interface (BMI)](https://bmi.readthedocs.io/en/latest/).

### These libraries should all be available in the environment.yml through the command
`conda activate bmi_lstm`

```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from netCDF4 import Dataset
import sys
sys.path.append("../src/")
import bmi_lstm # This is the BMI LSTM that we will be running from the file: bmi_lstm.py 
import pickle
```

### This sample dataset comes from NeuralHydrology: 
These are just samples. These models can run with any forcing data. This is just a sample that is used for examples within NeuralHydrology.
https://github.com/neuralhydrology/neuralhydrology/tree/master/test/test_data/camels_us/hourly
* 02064000 Falling River nr Naruna, VA
* 01547700 Marsh Creek at Blanchard, PA
* 03015500 Brokenstraw Creek at Youngsville, PA
* 01022500 Narraguagus River at Cherryfield, Maine

```python
sample_data = Dataset('../data/usgs-streamflow-nldas_hourly.nc', 'r')
sample_basins = {sample_data['basin'][x]:x for x in range(len(list(sample_data['basin'])))}
```

### Set a time constant
So that the time of the sample data aligns with the time of the NeuralHydrology evaluation. This will allow us to compare the implimentation of the LSTM through BMI with that of the training. There will be some discrepancy between the NeuralHydrology evaluation, because that uses a 336 hour lookback period, and our BMI implimentation does not reset the states. We will soon add the option to do a specific lookback period or not.

```python
t2000 = 168721
```

# Test the model with the limited input

```python
# Load in the evaluation results from the NeuralHydrology training directory. This is just for plotting
with open("../trained_neuralhydrology_models/hourly_all_forcings_lat_lon_elev/test_results_01022500_E5.p", "rb") as fb:
    eval_hourly_all_forcings_lat_lon_elev_01022500_E5 = pickle.load(fb)
nh_hourly_all_forcings_lat_lon_elev_01022500 = \
    eval_hourly_all_forcings_lat_lon_elev_01022500_E5['01022500']['1H']['xr']['QObs(mm/d)_sim'].data[:,-1]/24

# Create an instance of the LSTM model with BMI
model_all = bmi_lstm.bmi_LSTM()


# Initialize the model with a configuration file
model_all.initialize(bmi_cfg_file=Path('../bmi_config_files/01022500_hourly_forcings_lat_lon_elev.yml'))

# The basin listed in the configuration has to be included in the sample basins for this example to run
# This is the location of the example basin in the sample data provided
ibasin = sample_basins[model_all.cfg_bmi['basin_id']]
# Note that this model can run on any basin, but this example include observations from specific basins

# Run the model with sample data

# Store the output in a list, so we can plot it afterwards
runoff_output_list_all = []

# Loop through the sample data
for precip, temp, lw, sw, shum, pres, w_u, w_v in zip(list(sample_data['total_precipitation'][ibasin].data),
                        list(sample_data['temperature'][ibasin].data),
                        list(sample_data['longwave_radiation'][ibasin].data),
                        list(sample_data['shortwave_radiation'][ibasin].data),
                        list(sample_data['specific_humidity'][ibasin].data),
                        list(sample_data['pressure'][ibasin].data),
                        list(sample_data['wind_u'][ibasin].data),
                        list(sample_data['wind_v'][ibasin].data)):
    
    # Set the model forcings to those in the sample data
    model_all.set_value('atmosphere_water__time_integral_of_precipitation_mass_flux',np.atleast_1d(precip))
    model_all.set_value('land_surface_air__temperature',np.atleast_1d(temp))
    model_all.set_value('land_surface_radiation~incoming~longwave__energy_flux',np.atleast_1d(lw))
    model_all.set_value('land_surface_radiation~incoming~shortwave__energy_flux',np.atleast_1d(sw))
    model_all.set_value('atmosphere_air_water~vapor__relative_saturation',np.atleast_1d(shum))
    model_all.set_value('land_surface_air__pressure',np.atleast_1d(pres))
    model_all.set_value('land_surface_wind__x_component_of_velocity',np.atleast_1d(w_u))
    model_all.set_value('land_surface_wind__y_component_of_velocity',np.atleast_1d(w_v))

    # Updating the model calculates the runoff from the inputs and the model state at this time step
    model_all.update()
    
    # Add the output to a list so we can plot
    runoff_output_list_all.append(model_all.get_value('land_surface_water__runoff_depth'))
    
# Plot a portion of the output.
start_plot=17000
end_plot  =24000
plt.plot(sample_data['qobs_CAMELS_mm_per_hour'][ibasin][start_plot:end_plot], label='obs', c='k')
plt.plot(runoff_output_list_all[start_plot:end_plot],label='LSTM BMI')
plt.plot(nh_hourly_all_forcings_lat_lon_elev_01022500[t2000+start_plot:t2000+end_plot],label='LSTM NH')
#plt.ylim([0,0.5])
plt.ylabel('streamflow mm/hr')
plt.xlabel('hours')
plt.legend()
plt.show()
plt.close()

# Calculate a metric
obs = np.array(sample_data['qobs_CAMELS_mm_per_hour'][ibasin])
sim = np.array(runoff_output_list_all)
denominator = ((obs - obs.mean())**2).sum()
numerator = ((sim - obs)**2).sum()
value = 1 - numerator / denominator
print("NSE: {:.2f}".format(1 - numerator / denominator))
```

## Now, for the sake of demonstration we are gong to generate some synthetic data and use that to run a model

```python
# Generate some synthetic "precipitation" as input
num_records = 1000
fake_precip_list = [0]
for i in range(1,num_records):
    
    # some percent of time we have no rain at all
    if np.random.uniform(0,1) < 0.98:
        fake_precip_list.append(0)
        
    # When we do have rain, the probability of heavy or light rain depends on the previous day's rainfall
    else:
        
        # If yesterday was a light rainy day, or no rain, then we are likely to have light rain today
        if fake_precip_list[i-1] < 1:
            if np.random.uniform(0,1) < 0.7:
                fake_precip_list.append(np.random.uniform(0,1))
            else:
                # But if we do have heavy rain, then it could be very heavy
                fake_precip_list.append(np.random.uniform(0,8))
                
        # If it was heavy rain yesterday, then we are likely to have heavy rain again today
        else:
            if np.random.uniform(0,1) < 0.3:
                fake_precip_list.append(np.random.uniform(0,1))
            else:
                fake_precip_list.append(np.random.uniform(0,5))
fake_temp_list = [np.sin(i) for i in range(len(fake_precip_list))]
```

```python
# Create an instance of the LSTM model with BMI
model_fake = bmi_lstm.bmi_LSTM()


# Initialize the model with a configuration file
model_fake.initialize(bmi_cfg_file=Path('../bmi_config_files/fake_hourly_slope_mean_precip_temp.yml'))

# Run the model with sample data

# Store the output in a list, so we can plot it afterwards
runoff_output_list_fake = []

# Loop through the sample data
for precip, temp in zip(fake_precip_list,fake_temp_list):
    
    # Set the model forcings to those in the sample data
    model_fake.set_value('atmosphere_water__time_integral_of_precipitation_mass_flux',np.atleast_1d(precip))
    model_fake.set_value('land_surface_air__temperature',np.atleast_1d(temp))

    # Updating the model calculates the runoff from the inputs and the model state at this time step
    model_fake.update()

    # Add the output to a list so we can plot
    runoff_output_list_fake.append(model_fake.get_value('land_surface_water__runoff_depth'))
    

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(runoff_output_list_fake, label='LSTM BMI')
ax2.plot(fake_precip_list, label='fake precip', c='gray')
ax1.set_xlabel('Hours')
ax1.set_ylabel('Streamflow')
ax2.set_ylabel('precipitation')

plt.legend()
plt.show()
plt.close()
```
